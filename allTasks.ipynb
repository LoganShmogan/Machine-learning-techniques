# Task 1: Linear Regression - Stock Price Prediction

### Action I - Data Loading
Predicting Apple Inc. (AAPL) closing stock prices using features like Open, High, Low, and Volume.

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Get stock data
ticker = yf.Ticker("AAPL")
df = ticker.history(period="5y")
print("Sample data:")
print(df.head())

# Column Explanations:
# Date (index) – Trading day.
# Open – Price when the stock first traded that day.
# High – Highest price during that day.
# Low – Lowest price during that day.
# Close – Price at the end of the trading day (this is our target y).
# Volume – Number of shares traded that day.
# Dividends – Dividend issued per share on that day (if any).
# Stock Splits – Stock split ratio (if a split occurred that day).
```

### Action II - Feature Selection
Choosing Features (X) to Predict Close:
- Open → starting price of the day (strongly related to Close)
- High → daily max, often correlated with closing price
- Low → daily min, also correlated
- Volume → market activity, might influence price movement

Not ideal for prediction: Dividends (rare events, mostly 0), Stock Splits (rare events, mostly 0)

```python
# Define Features (X) and Target (y)
X = df[["Open", "High", "Low", "Volume"]]
y = df["Close"]

print("Features (X) sample:")
print(X.head())
print("\nTarget (y) sample:")
print(y.head())
```

### Action III - Visualization

```python
# Plot open and close values
plt.figure(figsize=(10, 6))
plt.scatter(df["Open"], df["Close"], alpha=0.5)
plt.xlabel("Open Price")
plt.ylabel("Close Price")
plt.title("Scatter Plot: Open vs Close (AAPL)")
plt.show()
```

### Action IV - Correlation Analysis

```python
# Correlation analysis
corr = df[["Open", "High", "Low", "Volume", "Close"]].corr()
print("Correlation with Close:")
print(corr["Close"])

# Observations:
# Open → Close: Very strong positive correlation (~0.99)
# High → Close: Very strong positive correlation (~0.99)
# Low → Close: Very strong positive correlation (~0.99)
# Volume → Close: Negative correlation (-0.525277)
# This suggests that higher trading activity sometimes coincides with price dropping.
```

### Action V - Train-Test Split

```python
# Split dataset: 70% training, 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

### Action VI - Model Training

```python
# Multivariate linear regression
model = LinearRegression()
model.fit(X_train, y_train)
print("Model trained successfully")
```

### Action VII - Model Parameters

```python
print("Intercept (b0):", model.intercept_)
print("Coefficients (b1, b2, ...):")
for feature, coef in zip(X.columns, model.coef_):
    print(f"  {feature}: {coef}")
```

### Action VIII - Model Evaluation

```python
# R^2 scores
train_r2 = model.score(X_train, y_train)
test_r2 = model.score(X_test, y_test)

print("R^2 Score (Training):", train_r2)
print("R^2 Score (Testing):", test_r2)
```

### Action IX - Predictions

```python
# Example imaginary feature sets (Open, High, Low, Volume)
imaginary_data = np.array([
    [150, 155, 149, 50000000],
    [200, 205, 195, 70000000],
    [300, 310, 290, 100000000]
])

predictions = model.predict(imaginary_data)
print("Predicted imaginary close values:")
for i, pred in enumerate(predictions):
    print(f"  Input {i+1}: {pred}")
```

### Action X - Model Without Intercept

```python
# Refit with intercept as false
model_no_intercept = LinearRegression(fit_intercept=False)
model_no_intercept.fit(X_train, y_train)

print("Coefficients with no intercept:")
for feature, coef in zip(X.columns, model_no_intercept.coef_):
    print(f"  {feature}: {coef}")
```

# Task 3: Data Classification - Breast Cancer Diagnosis

### Action I - Dataset Overview
Diagnostic data used to classify breast tumors as either benign (non-cancerous) or malignant (cancerous).

Key attributes include:
- Clump thickness
- Uniformity of cell size
- Uniformity of cell shape
- Marginal adhesion
- Single epithelial cell size
- Bare nuclei
- Bland chromatin
- Normal nucleoli
- Mitoses

The dataset is widely used for classification tasks in medical diagnostics.

### Action II - Data Loading and Preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# Load dataset from UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
columns = [
    "id", "clump_thickness", "uniformity_cell_size", "uniformity_cell_shape",
    "marginal_adhesion", "single_epithelial_cell_size", "bare_nuclei",
    "bland_chromatin", "normal_nucleoli", "mitoses", "class"
]
df_bc = pd.read_csv(url, names=columns)

# Replace missing values marked as '?' with NaN and drop rows with missing values
df_bc.replace('?', np.nan, inplace=True)
df_bc.dropna(inplace=True)

# Convert all columns to numeric
for col in df_bc.columns:
    df_bc[col] = pd.to_numeric(df_bc[col])

# Separate features and target
X_bc = df_bc.drop(columns=['id', 'class'])
y_bc = df_bc['class']

print("Dataset loaded. Shape:", X_bc.shape)
```

### Action III - Train-Test Split

```python
X_train_bc, X_test_bc, y_train_bc, y_test_bc = train_test_split(X_bc, y_bc, test_size=0.27, random_state=42)
print(f"Training set size: {X_train_bc.shape[0]}")
print(f"Testing set size: {X_test_bc.shape[0]}")
```

### Action IV - Feature Scaling

```python
scaler_bc = StandardScaler()
X_train_scaled = scaler_bc.fit_transform(X_train_bc)
X_test_scaled = scaler_bc.transform(X_test_bc)
print("Features standardized")
```

### Action V - Logistic Regression

```python
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_scaled, y_train_bc)
y_pred_logreg = logreg.predict(X_test_scaled)
acc_logreg = accuracy_score(y_test_bc, y_pred_logreg)
print(f"Logistic Regression Accuracy: {acc_logreg:.4f}")
```

### Action VI - K-Nearest Neighbors

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train_bc)
y_pred_knn = knn.predict(X_test_scaled)
acc_knn = accuracy_score(y_test_bc, y_pred_knn)
print(f"KNN Accuracy: {acc_knn:.4f}")
```

### Action VII - Model Comparison

```python
print(f"Logistic Regression vs KNN: {acc_logreg:.4f} vs {acc_knn:.4f}")
```

### Confusion Matrices and Visualization

```python
# Confusion matrix for Logistic Regression
cm_logreg = confusion_matrix(y_test_bc, y_pred_logreg)
disp_logreg = ConfusionMatrixDisplay(confusion_matrix=cm_logreg, display_labels=["Benign", "Malignant"])
disp_logreg.plot(cmap='Blues')
plt.title("Logistic Regression Confusion Matrix")
plt.show()

# Confusion matrix for KNN
cm_knn = confusion_matrix(y_test_bc, y_pred_knn)
disp_knn = ConfusionMatrixDisplay(confusion_matrix=cm_knn, display_labels=["Benign", "Malignant"])
disp_knn.plot(cmap='Greens')
plt.title("KNN Confusion Matrix")
plt.show()

# Bar plot comparing accuracies
plt.bar(["Logistic Regression", "KNN"], [acc_logreg, acc_knn], color=['blue', 'green'])
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.ylim(0.9, 1.0)
plt.show()
```

# Task 4: Principal Component Analysis - Wine Classification

### Action I - Dataset Overview
Contains chemical analysis results of wines grown in the same region of Italy but derived from three different cultivars (grape varieties). Used for classification, clustering, and dimensionality reduction experiments.

Key attributes include:
- Alcohol
- Malic acid
- Ash
- Alcalinity of ash
- Magnesium
- Total phenols
- Flavanoids
- Nonflavanoid phenols
- Proanthocyanins
- Color intensity
- Hue
- OD280/OD315 of diluted wines
- Proline

### Action II - Data Loading and Preprocessing

```python
from sklearn.decomposition import PCA

# Load wine dataset
url_wine = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
columns_wine = [
    "Class", "Alcohol", "Malic_acid", "Ash", "Alcalinity_of_ash", "Magnesium",
    "Total_phenols", "Flavanoids", "Nonflavanoid_phenols", "Proanthocyanins",
    "Color_intensity", "Hue", "OD280/OD315", "Proline"
]
df_wine = pd.read_csv(url_wine, header=None, names=columns_wine)

X_wine = df_wine.drop(columns=["Class"])
y_wine = df_wine["Class"]

# Standardize features
scaler_wine = StandardScaler()
X_wine_scaled = scaler_wine.fit_transform(X_wine)

print(f"Dataset loaded. Shape: {X_wine.shape}")
print(f"Number of classes: {len(np.unique(y_wine))}")
```

### Action III - PCA Transformation

```python
pca = PCA(n_components=2)
X_wine_pca = pca.fit_transform(X_wine_scaled)
print(f"PCA completed. Explained variance ratio: {pca.explained_variance_ratio_}")
```

### Action IV - Classification on PCA Features

```python
X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(X_wine_pca, y_wine, test_size=0.3, random_state=42)
logreg_wine = LogisticRegression(max_iter=1000)
logreg_wine.fit(X_train_wine, y_train_wine)
y_pred_wine = logreg_wine.predict(X_test_wine)
acc_wine = accuracy_score(y_test_wine, y_pred_wine)
print(f"Logistic Regression Accuracy (PCA features): {acc_wine:.4f}")
```

### Action V - PCA Visualization

```python
plt.figure(figsize=(10, 8))
for class_value in np.unique(y_wine):
    plt.scatter(
        X_wine_pca[y_wine == class_value, 0],
        X_wine_pca[y_wine == class_value, 1],
        label=f"Class {class_value}",
        alpha=0.7
    )
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Wine Dataset: PCA Feature Scatter Plot")
plt.legend()
plt.show()
```
